```python
import pandas as pd
df = pd.read_csv('/content/Hashtag_sum_2_10.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Chuyển đổi 'date' thành datetime
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Lọc ra các cột hashtag để tính tương quan
hashtag_columns = [col for col in df.columns if col.startswith('#')]

# Tính ma trận tương quan
correlation_matrix = df[hashtag_columns].corr()

# Vẽ heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Ma trận tương quan giữa các Hashtag", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(7, 1, figsize=(14, 24))

# Biểu đồ #covid19
axes[0].plot(df['date'], df['#covid19'], label='#covid19', alpha=0.7)
axes[0].set_title("", fontsize=14)
axes[0].set_xlabel("")
axes[0].set_ylabel("")
axes[0].grid(True)
axes[0].legend()

# Biểu đồ #crypto va #bitcoin
axes[1].plot(df['date'], df['#bitcoin'], label='#bitcoin', alpha=0.7)
axes[1].plot(df['date'], df['#crypto'], label='#crypto', alpha=0.7)
axes[1].set_title("", fontsize=14)
axes[1].set_xlabel("")
axes[1].set_ylabel("")
axes[1].grid(True)
axes[1].legend()


# Biểu đồ #ukraine va #iran
axes[2].plot(df['date'], df['#iran'], label='#iran', alpha=0.7)
axes[2].plot(df['date'], df['#ukraine'], label='#ukraine', alpha=0.7)
axes[2].set_title("", fontsize=14)
axes[2].set_xlabel("")
axes[2].set_ylabel("")
axes[2].grid(True)
axes[2].legend()

# Biểu đồ #christmas
axes[3].plot(df['date'], df['#christmas'], label='#christmas', alpha=0.7)
axes[3].set_title("", fontsize=14)
axes[3].set_xlabel("")
axes[3].set_ylabel("")
axes[3].grid(True)
axes[3].legend()

# Biểu đồ #travel
axes[4].plot(df['date'], df['#travel'], label='#travel', alpha=0.7)
axes[4].set_title("", fontsize=14)
axes[4].set_xlabel("")
axes[4].set_ylabel("")
axes[4].grid(True)
axes[4].legend()

# Biểu đồ #mondaymotivation
axes[5].plot(df['date'], df['#mondaymotivation'], label='#mondaymotivation', alpha=0.7)
axes[5].set_title("", fontsize=14)
axes[5].set_xlabel("")
axes[5].set_ylabel("")
axes[5].grid(True)
axes[5].legend()

# Biểu đồ #realestate
axes[6].plot(df['date'], df['#realestate'], label='#realestate', alpha=0.7)
axes[6].set_title("", fontsize=14)
axes[6].set_xlabel("")
axes[6].set_ylabel("")
axes[6].grid(True)
axes[6].legend()

plt.tight_layout()
plt.show()
```

```python
# Chuyển cột 'date' thành kiểu datetime
df['date'] = pd.to_datetime(df['date'], format="%d/%m/%Y")

# Tạo cột 'month' để lấy tháng từ cột 'date'
df['month'] = df['date'].dt.month
# Tạo biểu đồ mùa vụ theo tháng cho các hashtag với yêu cầu gộp
plt.figure(figsize=(14, 24))


# Vẽ biểu đồ #covid19 theo tháng
plt.subplot(7, 1, 1)
df.groupby('month')['#covid19'].mean().plot(label='#covid19', marker='o')
plt.title("", fontsize=14)
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.legend()

# Gộp #crypto và #bitcoin vào 1 biểu đồ
plt.subplot(7, 1, 2)
df.groupby('month')['#crypto'].mean().plot(label='#crypto', marker='o')
df.groupby('month')['#bitcoin'].mean().plot(label='#bitcoin', marker='o')
plt.title("", fontsize=14)
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.legend()

# Gộp #ukraine và #iran vào 1 biểu đồ
plt.subplot(7, 1, 3)
df.groupby('month')['#ukraine'].mean().plot(label='#ukraine', marker='o')
df.groupby('month')['#iran'].mean().plot(label='#iran', marker='o')
plt.title("", fontsize=14)
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.legend()

# Vẽ biểu đồ #mondaymotivation theo tháng
plt.subplot(7, 1, 4)
df.groupby('month')['#mondaymotivation'].mean().plot(label='#mondaymotivation', marker='o')
plt.title("", fontsize=14)
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.legend()

# Vẽ biểu đồ #realestate theo tháng
plt.subplot(7, 1, 5)
df.groupby('month')['#realestate'].mean().plot(label='#realestate', marker='o')
plt.title("", fontsize=14)
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.legend()

# Vẽ biểu đồ #christmas theo tháng
plt.subplot(7, 1, 6)
df.groupby('month')['#christmas'].mean().plot(label='#christmas', marker='o')
plt.title("", fontsize=14)
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.legend()

# Vẽ biểu đồ #travel theo tháng
plt.subplot(7, 1, 7)
df.groupby('month')['#travel'].mean().plot(label='#travel', marker='o')
plt.title("", fontsize=14)
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc lại dữ liệu từ file
file_path = '/content/Hashtag_sum_2_10.csv'
df = pd.read_csv(file_path)

# Chuyển cột 'date' thành kiểu datetime
df['date'] = pd.to_datetime(df['date'], format="%d/%m/%Y")

# Tạo cột ngày trong tuần (0: Thứ 2, 6: Chủ nhật)
df['day_of_week'] = df['date'].dt.dayofweek

# Tạo biểu đồ mùa vụ theo ngày trong tuần cho #travel và #mondaymotivation
plt.figure(figsize=(14, 6))

# Mùa vụ #travel theo ngày trong tuần
plt.subplot(2, 1, 1)
df.groupby('day_of_week')['#travel'].mean().plot(label='#travel', marker='o')
plt.title("", fontsize=14)
plt.xlabel("")
plt.ylabel("")
plt.xticks([0, 1, 2, 3, 4, 5, 6], ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.grid(True)
plt.legend()

# Mùa vụ #mondaymotivation theo ngày trong tuần
plt.subplot(2, 1, 2)
df.groupby('day_of_week')['#mondaymotivation'].mean().plot(label='#mondaymotivation', marker='o')
plt.title("", fontsize=14)
plt.xlabel("")
plt.ylabel("")
plt.xticks([0, 1, 2, 3, 4, 5, 6], ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
